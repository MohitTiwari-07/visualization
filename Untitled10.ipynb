```python
from sklearn.datasets import load_files

data_dir = "/content/20_newsgroups_data/20_newsgroups"
data = load_files(data_dir, encoding='latin1', decode_error='ignore')
documents = data.data
target_names = data.target_names
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import TreebankWordTokenizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))
tokenizer = TreebankWordTokenizer()  # No downloads needed!

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = tokenizer.tokenize(text)
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words and len(w) > 2]
    return " ".join(words)

# Make sure 'documents' is defined (rerun your load_files() cell if needed)
clean_docs = [preprocess(doc) for doc in documents]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

tfidf = TfidfVectorizer(max_df=0.5, min_df=5, stop_words='english')
X_tfidf = tfidf.fit_transform(clean_docs)

kmeans = KMeans(n_clusters=20, random_state=42)
labels = kmeans.fit_predict(X_tfidf)

score = silhouette_score(X_tfidf, labels)
print(f"✅ Silhouette Score (KMeans): {score:.3f}")
```

```python
def top_kmeans_terms(model, vectorizer, n_terms=10):
    terms = vectorizer.get_feature_names_out()
    for i, center in enumerate(model.cluster_centers_):
        top_terms = center.argsort()[-n_terms:][::-1]
        print(f"Cluster {i}: " + " ".join(terms[top_terms]))

top_kmeans_terms(kmeans, tfidf)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

count = CountVectorizer(max_df=0.5, min_df=5, stop_words='english')
X_count = count.fit_transform(clean_docs)

lda = LatentDirichletAllocation(n_components=20, max_iter=10,
                                 learning_method='online', random_state=42)
lda.fit(X_count)
```

```python
def top_lda_topics(model, vectorizer, n_terms=10):
    terms = vectorizer.get_feature_names_out()
    for idx, topic in enumerate(model.components_):
        top = topic.argsort()[-n_terms:][::-1]
        print(f"Topic {idx}: " + " ".join([terms[i] for i in top]))

top_lda_topics(lda, count)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

for idx, topic in enumerate(lda.components_):
    plt.figure(figsize=(8, 5))
    words = {
        count.get_feature_names_out()[i]: topic[i]
        for i in topic.argsort()[-30:]
    }
    wc = WordCloud(background_color='white').generate_from_frequencies(words)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"Topic {idx}")
    plt.show()
```

```python
import pandas as pd
topics = lda.transform(X_count)
df = pd.DataFrame(topics, columns=[f"Topic {i}" for i in range(20)])
df.to_csv("lda_topic_distribution.csv", index=False)
```
